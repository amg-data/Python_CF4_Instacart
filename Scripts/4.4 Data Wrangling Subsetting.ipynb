```python
#dropping columns - remove eval_set column from df_ords visually
df_ords.drop(columns = ['eval_set'])
```

```python
#transposing data - rows to columns to rows - wide format to long format

#import departments data
df_dep = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'departments.csv'), index_col = False)

#check df_dep
df_dep.head()
```

#05. Subsetting - creating a smaller data set from the whole set using a filter

```python
#create filtered df (df_snacks) using department_id 19 in the df_prods dataframe
df_snacks =  df_prods[df_prods['department_id']==19]

#check df_snacks
df_snacks.head()
```

```python
#create filtered df (df_snacks_2) using loc function
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]

#check df_snacks
df_snacks_2.head()
```

```python
#create filtered df (df_snacks_3) using isin function - allows for checking multiple id numbers [17,18,19]
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]

#check df_snacks
df_snacks_3.head()
```
